### Plotting the Data

```python
#Instalar matplotlib
%pip install matplotlib  
import matplotlib.pyplot as plt
```

```python
#Grafica los precios de la clumna S&P500
df_comp.spx.plot(figsize=(20,5), title = "S&P500 Prices")
plt.show()
```

```python
#Grafica los precios de la clumna FTSE100
df_comp.ftse.plot(figsize=(20,5), title = "FTSE100 Prices")
plt.show()
```

```python
#La mezcla de los dos anteriores graficos
df_comp.spx.plot(figsize=(20,5), title = "S&P500 Prices")
df_comp.ftse.plot(figsize=(20,5), title = "FTSE100 Prices")
plt.title("S&P v FTSE")
plt.show()
```

### The QQ Plot

```python
#Biblioteca SciPy 
%pip install scipy 
import scipy.stats #Amplia gama de funciones y distribuciones estadísticas 
import pylab #Computación científica y la creación de gráficos.
```

```python
#Gráfico de probabilidad (Q-Q plot) para evaluar si una muestra de datos sigue una distribución teórica específica
scipy.stats.probplot(df_comp.spx,plot = pylab)
plt.title("QQ Plot", size = 24)
pylab.show()
```
