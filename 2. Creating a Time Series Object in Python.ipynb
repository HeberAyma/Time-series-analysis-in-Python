### From Text to Date

```python
#Convertir una columna de fechas en un DataFrame al formato de fecha y hora de pandas
#La opción dayfirst=True se utiliza para indicar que el formato de las fechas tiene el día antes del mes en el formato de fecha.
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = False) 
```

```python
df_comp.date.describe() #Generar un resumen estadístico de la columna "date" en el DataFrame
```

### Setting the Index

```python
df_comp.set_index("date") #Establecer la columna "date" como el índice del DataFrame
```

```python
#inplace=True asegura que los cambios se realicen en el propio DataFrame en lugar de crear una nueva copia con el índice modificado
df_comp.set_index("date", inplace=True) 
```

```python
df_comp.head() #Muestra las primeras filas de la data
```

### Setting the Desired Frequency

```python
#Rellenar un DataFrame de pandas con un índice de fechas (DatetimeIndex) para asegurarse de que haya una fila para 
#cada día en el rango de fechas, incluso si no hay datos disponibles para algunos de esos días. 
#La frecuencia 'd' indica que deseas que se cree una fila para cada día
df_comp=df_comp.asfreq('d')
```

```python
df_comp #Mostrar los datos
```

```python
df_comp=df_comp.asfreq('b') #Cambiar la frecuencia de muestreo de un DataFrame de pandas a días laborables
```

### Handling Missing Values

```python
#Muestra valor faltante 
#True = hay valor faltante y False = no hay valor faltante
df_comp.isna()
```

```python
#Obtener un número que representa la cantidad de valores faltantes
df_comp.isna().sum()
```

```python
df_comp #Mostrar los datos
```

```python
df_comp.spx=df_comp.spx.fillna(method='ffill') #Metodo de llenado de datos hacia adelante para la columna spx
```

```python
df_comp.isna().sum() ##Obtener un número que representa la cantidad de valores faltantes
```

```python
df_comp.ftse=df_comp.ftse.fillna(method='bfill') #Metodo de llenado de datos hacia atras para la columna ftse
```

```python
df_comp.dax=df_comp.dax.fillna(value = df_comp.dax.mean()) #Reemplazar los valores faltantes de "dax" con el promedio de "dax"
```
